# 1. Write a Python Program to Find the Factorial of a Number?

```python
def factorial(n):
    result =1
    for i in range(2,n+1):
        result*=i
    return result
num= int(input("Enter a non negative number: "))
if n<0:
    print("result is invalid")
else:
    result = factorial(num)
    print(f"The factorial of {num}is {result}")
```

```python
factorial(-1)
```

# 2. Write a Python Program to Display the multiplication Table?

```python
def multitable(n):
    for i in range(1,n+1):
        print(f"mutiplication table for {i}")
        for j in range(1,11):
            result = i*j
            
            print(f"{i}*{j} = {result}")
        print("----------")
multitable(30)
```

# 3. Write a Python Program to Print the Fibonacci sequence?

```python
nth = int(input("Enter fib_series terms : "))
n1=0
n2=1
count=0
if nth<0:
    print(n1)
elif nth==1:
    print(n2)
else:
    while count<nth:
        print(n1)
        n =n1+n2
        n1=n2
        n2=n
        count+=1
        
```

# 4. Write a Python Program to Check Armstrong Number?

```python
def checkArmst(num):
    num_str =str(num)
    n=len(num_str)
    sum=0
    for digits in num_str:
        sum += int(digits)**n
        
    if sum==num:
        return True
    else:
        return False
#num=371
```

```python
checkArmst(371)
```

# 5. Write a Python Program to Find Armstrong Number in an Interval?

```python
lower_bound = int(input("Enter the lower limit :"))
upper_bound = int(input("Enter the upper limit :"))

for num in range(lower_bound,upper_bound):
    order = len(str(num))
    sum = 0
    temp =num
    while temp>0:
        digit = temp%10
        sum += digit**order
        temp //=10
    if num==sum:
        print(num)
```

# 6. Write a Python Program to Find the Sum of Natural Numbers?

```python
def sumNatural(num):
    if num==0:
        return 0
    else:
        return num+sumNatural(num-1)
sumNatural(100)
```
